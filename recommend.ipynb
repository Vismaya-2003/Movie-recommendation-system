Data Pre-processing the merged dataset

```python
# Displaying the first instance of the movies dataset
movies.head(1)
```

```python
# Checking how the genre column looks
movies.iloc[0].genres
```

```python
# Converts a string representation of a list of dictionaries into an actual list of dictionaries
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# Converting genre column into a list containing genre names
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
# Applies the 'convert' function to each element in the 'genres' column of the 'movies' DataFrame
movies['genres']=movies['genres'].apply(convert)
```

```python
# Displaying the first five instances of the movies dataset after genre column modification
movies.head()
```

```python
# Applies the 'convert' function to each element in the 'keywords' column of the 'movies' DataFrame
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# Displaying the first five instances of the movies dataset after keyword column modification
movies.head()
```

```python
# Checking how the cast column looks
movies.iloc[0].cast
```

```python
# Extracting the first three dictionaries from the cast column
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3: 
          L.append(i['name'])
          counter += 1
        else:
            break
    return L    
```

```python
# Applies the 'convert3' function to each element in the 'cast' column of the 'movies' DataFrame
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# Displaying the first five instances of the movies dataset after cast column modification
movies.head()
```

```python
# Checking how the crew column looks
movies.iloc[0].crew
```

```python
# Fetching the director name from the crew column of the movies dataset
def fetch_diector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
          L.append(i['name'])
          break
    return L    
```

```python
# Applies the 'fetch_director' function to each element in the 'crew' column of the 'movies' DataFrame
movies['crew'] = movies['crew'].apply(fetch_diector)
```

```python
# Displaying the first five instances of the movies dataset after crew column modification
movies.head()
```

```python
# Converting the overview column into a string
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Displaying the first five instances of the movies dataset after data type conversion
movies.head()
```

```python
# Removes all spaces from each string in the 'genres', 'keywords', 'cast', and 'crew' columns of the 'movies' DataFrame
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

Performing Stemming on the dataset

```python
import nltk # For stemming
```

```python
from nltk.stem.porter import PorterStemmer # Importing library from nltk

# Intitializing Porter stemmer
ps = PorterStemmer()
```

```python
# Stemming
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# Applying stemming on tags
movie_new['tags']= movie_new['tags'].apply(stem)
```

```python
# Checking if stemmimg happen
ps.stem("loving")
```

```python
# Checking if the stemming function works
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

Performing Vectorization on the movie_new dataset

```python
# Importing necessary library for vectorization process
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Displaying the first five instances of the movie_new dataset
movie_new.head()
```

```python
# Initialize CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fit and transform the data
vectors = cv.fit_transform(movie_new['tags'])

# Get feature names
feature_names = cv.get_feature_names_out()
```

```python
# Getting the shape of the vector
vectors.shape
```

```python
# Displaying the vector matrix
vectors = vectors.toarray()
print(vectors)
```

```python
# Displaying the vectors of the first movie
vectors[0]
```

```python
# Print all feature names (after stemming)
for name in feature_names:
    print(name)
```

```python
# Import the cosine_similarity function to compute similarity between vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Compute the cosine similarity matrix for the given vectors
similarity = cosine_similarity(vectors)
```

```python
# Displaying the similarity matrix
similarity
```

```python
# Displaying the shape of the similarity matrix
similarity.shape
```

```python
# Checking the similarity index of the first movie with evry other movie in the dataset
similarity[0]
```

```python
# Enumerate the similarity scores, sort them in descending order by the score, and retrieve the top 5 most similar items (excluding the first item)
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
# Sort the similarity values of the first item in descending order
sorted(similarity[0], reverse=True)
```

```python
# Retrieve the index of the movie 'Batman Begins' in the DataFrame
movie_new[movie_new['title']== 'Batman Begins'].index[0]
```

Building the recommendation model

```python
def recommend(movie):
    # Find the index of the movie in the DataFrame
    movie_index = movie_new[movie_new['title'] == movie].index[0]
    
    # Get the similarity scores for the movie at `movie_index`
    distances = similarity[movie_index]
    
    # Enumerate, sort, and get the top 5 most similar movies (excluding the movie itself)
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print the indices of the top 5 recommended movies
    for i in movies_list:
        print(movie_new.iloc[i[0]].title)
```

```python
# Call the recommend function to get and print the indices of the top 5 movies similar to 'Avatar'
recommend('Avatar')
```

```python
# Testing if the model works properly
recommend('Batman Begins')
```
